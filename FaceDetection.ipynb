```python
def convertToRGB(img):
    return cv2.cvtColor(img, cv2.COLOR_BGR2RGB);
test1 = cv2.imread('data/testbaby.jpg');
gray_img = cv2.cvtColor(test1,cv2.COLOR_BGR2GRAY);
plt.imshow(gray_img, cmap='gray');
```

```python
haar_face_cascade = cv2.CascadeClassifier('data/haarcascade_frontalface_alt.xml');
faces = haar_face_cascade.detectMultiScale(gray_img, scaleFactor=1.1, minNeighbors=5);
print("NO OF FACES",faces);
for (x, y, w, h) in faces:     
         cv2.rectangle(test1, (x, y), (x+w, y+h), (0, 255, 0), 2);
        

plt.imshow(convertToRGB(test1));
```

```python


img=cv2.imread('data/test4.jpg');

#for Haar_CAscade classifier
haar_classifier=cv2.CascadeClassifier('data/haarcascade_frontalface_alt.xml');

#for LBP classifier
lbp_classifier=cv2.CascadeClassifier('data/lbpcascade_frontalface.xml');


def convertToRGB(img):
    return cv2.cvtColor(img,cv2.COLOR_BGR2RGB);

def faceDetect(img,classifier,scaleFactor):
    img_copy=img.copy();
    gray_img_copy=cv2.cvtColor(img_copy,cv2.COLOR_BGR2GRAY);
    faces=classifier.detectMultiScale(gray_img_copy,scaleFactor,minNeighbors=5);
    print("FO of faces found : ",len(faces));
    for (x, y, w, h) in faces:
      cv2.rectangle(img_copy, (x,y), (x+w, y+h), (255, 0, 0), 2);
    
    return img_copy;



t1=time.time();
haar_afterDetection = faceDetect(img,haar_classifier,1.2);
t2=time.time();
dt1=t2-t1;

print("Time for HAAR :",dt1);

#plt.imshow(convertToRGB(haar_afterDetection));

#time.sleep(5);

t1=time.time();
lbp_afterDetection = faceDetect(img,lbp_classifier,1.1);
t2=time.time();
dt2=t2-t1;

print("Time for lbp :",dt2);

f, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5)) 
 
 #show Haar image 
ax1.set_title('Haar Detection time: ' + str(round(dt1, 3)) + ' secs') 
ax1.imshow(convertToRGB(haar_afterDetection))  
 
#show LBP image 
ax2.set_title('LBP Detection time: ' + str(round(dt2, 3)) + ' secs') 
ax2.imshow(convertToRGB(lbp_afterDetection)) 


    
```
